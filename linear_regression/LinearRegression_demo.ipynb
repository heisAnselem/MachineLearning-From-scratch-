# Univariate Linear Regression (from Scratch)

This notebook demonstrates my custom-built **LinearRegression** class
using generated synthetic data.  

The goal is to show how gradient descent can recover the slope and intercept of a line.

```python
import numpy as np
import matplotlib.pyplot as plt
from Univariate_model import LinearRegression
```

## Generate Synthetic Data
creating data that follows the line:

\[
y = 3x + 7 + noise in data
\]

This way, we know the "true" slope (3) and intercept (7), 
and can check if our model learns them correctly.

```python
# Creating  synthetic data
np.random.seed(42)
X = np.linspace(0, 10, 100)
y = 3 * X + 7 + np.random.randn(100) * 2   # adds Gaussian noise

# Quick plot of the data just to visualize 
plt.scatter(X, y, alpha=0.7)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Generated Data (Linear Relation + Noise)")
plt.show()
```

## Training the Model
I now train the `LinearRegression` model using the fit method (gradient descent under the hood).

```python
# Initializing and fitting the data
model = LinearRegression(learning_rate=0.01,epochs=1000,echoe=True,tolerance=1e-4)
model.fit(X,y)
```

## Visualize the Regression Line
Plotting the fitted line against the noisy data

```python
# predictions
y_pred = model.predict(X)

# Plot
plt.figure(figsize=(8,6))
plt.scatter(X, y, label="Data (with noise)", alpha=0.7)
plt.plot(X, y_pred, color="red", linewidth=2, label="Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Univariate Linear Regression from Scratch")
plt.legend()
plt.show()
```

## Evaluating  the Model
i'll use **Mean Squared Error (MSE)** and **Root Mean Sqaured Error (RMSE)** to evaluate the fit.

```python
mse = model.mean_squared_error(X, y)
rmse = model.root_mean_squared_error(X,y)
print(
    f"Mean Squared Error: {mse:.4f}\n"
    f"Root Mean Squared Error : {rmse:.4f}"
    )
```

The model actually performed well given that the difference between our target  weights and biases is very little
